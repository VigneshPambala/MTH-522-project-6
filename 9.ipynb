```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Load the CSV file
file_name = "input2.csv"  # Replace with your file name
data = pd.read_csv(file_name)

# Ensure the column names are appropriate
# Replace 'x', 'y', and 'label' with your actual column names if they differ
x = data['x']
y = data['y']
labels = data['label']

# Calculate z = sqrt(x^2 + y^2)
z = np.sqrt(x**2 + y**2)

# Add z to the DataFrame
data['z'] = z

# Separate points by label
class_neg1 = data[labels == -1]
class_pos1 = data[labels == 1]

# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot points for each class
ax.scatter(class_neg1['x'], class_neg1['y'], class_neg1['z'], color='blue', label='Label -1')
ax.scatter(class_pos1['x'], class_pos1['y'], class_pos1['z'], color='red', label='Label 1')

# Customize the plot
ax.set_title('3D Scatter Plot of Points in R^3')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z (distance from origin)')
ax.legend()
plt.show()
```
