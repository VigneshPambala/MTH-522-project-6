```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
# Replace 'your_file.csv' with the actual file path
file_path = 'input.csv'
data = pd.read_csv(file_path)

# Extract columns
x = data.iloc[:, 0]  # First column: x-axis positions
y = data.iloc[:, 1]  # Second column: y-axis positions
labels = data.iloc[:, 2]  # Third column: -1 or 1 values

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x[labels == -1], y[labels == -1], color='blue', label='Label -1')
plt.scatter(x[labels == 1], y[labels == 1], color='red', label='Label 1')

# Add plot details
plt.xlabel('X-axis Position')
plt.ylabel('Y-axis Position')
plt.title('Scatter Plot of Points')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
