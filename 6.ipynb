```python
import pandas as pd

# Load the Excel file
# Replace 'your_file.xlsx' with the path to your Excel file
excel_file = 'input2.xlsx'
csv_file = 'input2.csv'  # Specify the desired output CSV file name

# Read the first sheet of the Excel file
data = pd.read_excel(excel_file, sheet_name=0)

# Save to a CSV file
data.to_csv(csv_file, index=False)

print(f"Excel file '{excel_file}' has been converted to CSV file '{csv_file}'.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_name = "input2.csv"  # Replace with your file name
data = pd.read_csv(file_name)

# Ensure the column names are appropriate
# Replace 'x', 'y', and 'label' with your actual column names if they differ
x = data['x']
y = data['y']
labels = data['label']

# Split the points by label
x_neg1 = x[labels == -1]
y_neg1 = y[labels == -1]

x_pos1 = x[labels == 1]
y_pos1 = y[labels == 1]

# Plot the points
plt.figure(figsize=(8, 6))
plt.scatter(x_neg1, y_neg1, color='blue', label='Label -1')
plt.scatter(x_pos1, y_pos1, color='red', label='Label 1')

# Customize the plot
plt.title('Scatter Plot of Points with Labels')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
